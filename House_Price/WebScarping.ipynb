```python
import pandas as pd
import numpy as np
import openpyxl
import requests
from bs4 import BeautifulSoup
```

```python
def url_gonder():
    url="https://www.emlakjet.com/kiralik-konut/istanbul/"
    url_list=[]
    for i in range(1,51):
        if(i==1):
            url_list.append(url)
        else:
            url=url.replace('"', '', 1)
            url_list.append(url+str(i)+'/')
        i+=1
    return url_list
```

```python
def url_al(url_list):
    ilan_adlari_listesi=[]
    ilan_fiyatlari_listesi=[]
    adres_bilgisi_listesi=[]
    m2_bilgisi_listesi=[]
    konut_tipi_listesi=[]
    ilan_tarihi_listesi=[]
    oda_sayisi_listesi=[]
    for url in url_list:
        sayfa=requests.get(url,headers={"HTTP_HOST":"Mozilla/5.0 (iPhone; CPU iPhone OS 17_3_1 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.3.1 Mobile/15E148 Safari/604.1"})
        soup=BeautifulSoup(sayfa.content,"html.parser")
        
        ilan_adlari=soup.find_all("div",{"class":"manJWF"})
        for ilan_adi in ilan_adlari:
            ilan_adlari_listesi.append(ilan_adi.find("h3").text)

        ilan_fiyatlari=soup.find_all("p",{"class":"_2C5UCT"})
        for ilan_fiyati in ilan_fiyatlari:
            ilan_fiyatlari_listesi.append(ilan_fiyati.find("span").text)

        adres_bilgileri=soup.find_all("div",{"class":"_2wVG12"})
        for adres_bilgisi in adres_bilgileri:
            adres_bilgisi_listesi.append(adres_bilgisi.find("span").text)

        m2_konutTipi_ilanTarihi_odaSayisi_Bilgisi=soup.find_all("div",{"class":"_2UELHn"})
        index=0
        for bilgiler in m2_konutTipi_ilanTarihi_odaSayisi_Bilgisi:
            for bilgi in bilgiler:
                if("weekend" in bilgi.text):
                    oda_sayisi_listesi.append(bilgi.text.replace("weekend",""))
                if("home" in bilgi.text):
                    konut_tipi_listesi.append(bilgi.text.replace("home",""))
                if("texture" in bilgi.text):
                    m2_bilgisi_listesi.append(bilgi.text.replace("texture",""))
                if("event" in bilgi.text):
                    ilan_tarihi_listesi.append(bilgi.text.replace("event",""))
    
      
    df = pd.DataFrame({'İlan Adı': ilan_adlari_listesi,
                       'İlan Fiyatı': ilan_fiyatlari_listesi,
                       'Adres bilgisi': adres_bilgisi_listesi,
                       'Konut büyüklüğü': m2_bilgisi_listesi,
                       'Konut tipi': konut_tipi_listesi,
                       'İlan''ın koyulma tarihi': ilan_tarihi_listesi,
                       'Oda sayısı': oda_sayisi_listesi,
                       })
    return df
```

```python
def excel_kayıt(df):
    df.to_excel('istanbul_konut_bilgileri.xlsx', index=False)
```

```python
url_list=url_gonder()
df=url_al(url_list)
df
```
